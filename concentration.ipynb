```python
print("n個の物質のなかで比較 : ")
output =int( input())

print("memory が持っているbit数 : ")
bit = int(input())
x_def = [] #inputのリストを作る
count1 = 1
while count1 <= bit  :
    
    print("X" , count1, "の濃度を入力 : ")  #化学反応式のXiの濃度
    xi = float(input())
    x_def.append(xi)  #xのリストに追加
    count1 += 1
    
print(x_def)

w = [] #重みのリストを作る    
weightlist = [] #output分のwを格納するリスト
count2 = 1
while count2 <= output :
    
    count1 = 1
    while count1 <= bit :  #bit回繰り返す
        print("W",count1, count2, "の濃度を入力 : ")
        wij = float(input())
        w.append(wij)
        count1 += 1
    weightlist.append(w) #できたリストwをweightlist に追加して値を保存する
    count2 += 1

print(weightlist)

count1 = 1 
xfi = [] #XFiの濃度を格納するリスト
while count1 <= bit :
    
    print("X",count1,"に対応するXF",count1,"の濃度を入力 : ")
    x = input()
    #x = x.replace('"','')
    x = float(x)              
    xfi.append(x)
    count1 += 1
    
print(xfi)

sgj = [] #SGjの濃度を格納するリスト
count2 = 1
while count2 <= output :
    
    print("y",count2,"に対応するSG",count2,"の濃度を入力 : ")
    x = float(input())
    sgj.append(x)
    count2 += 1
    
print(sgj)

rgj = []    #RGjの濃度を格納するリスト
count2 = 1    
while count2 <= output :
    
    print("y",count2,"に対応するRG",count2,"の濃度を入力 : ")
    x = float(input())
    rgj.append(x)
    count2 += 1    

print(rgj)

yfj = []    #YFjの濃度を格納するリスト
count2 = 1    
while count2 <= output :
    
    print("y",count2,"に対応するYF",count2,"の濃度を入力 : ")
    x =float( input())
    yfj.append(x)
    count2 += 1        
    
print(yfj)

repj = []    #Repjの濃度を格納するリスト
count2 = 1    
while count2 <= output :
    
    print("y",count2,"に対応するRep",count2,"の濃度を入力 : ")
    x = float(input())
    repj.append(x)
    count2 += 1     
print(repj)    
```

```python
print("Weight multiplication の速度定数ka : ")
x = float(input())
ka = x

print("Summation の速度定数kb : ")
x = float(input())
kb = x

print("Pairwise annihilation の速度定数kf : ")
x = float(input())
kf = x

print("Signal restoration の速度定数ks : ")
x = float(input())
ks = x

print("Reporting の速度定数kc : ")
x = float(input())
kc = x

print("Weight multiplication の実行時間 : ")
x = float(input())
t1 = x

print("Summation の実行時間 : ")
x = float(input())
t2 = x

print("Pairwise annihilation の実行時間 : ")
x = float(input())
t3 = x

print("Signal restoration の実行時間 : ")
x = float(input())
t4 = x

print("Reporting の実行時間 : ")
x = float(input())
t5 = x

compare = [] #pair annihilationで比べるためにSjを格納する
count2 = 1
while count2 <= output :    
#各jに対応する[Sj]を求める
    sum = 0 #Sjの合計の濃度
    count1 = 1
    while count1 <= bit :  #各XiによるSjを足してyjに対応する値にする
        pij = ka* x_def[count1-1] * xfi[count1-1] * weightlist[count2-1][count1-1] * t1
        sj = kb * pij * sgj[count2-1] * t2
        sum += sj
        count1 += 1
    compare.append(sum)  
    count2 += 1
    
print(compare)

#一番大きなSjのjの値を調べる   
a = 0
for i in range( 0, output-1 ):    
    if(compare[a] <= compare[i] ) :
        a = i
print(compare[a])

sumx= 0  #X(i+1)   のSignalrestorationで反応する濃度
sum1 = 0 
sum2 = 0
for l in range (0, a-1) :
    sj = (ks * compare[l] )/(ks + kf) 
    x = compare[a] - (compare[l]-sj) #SkとSjは１対１の濃度で反応するから
    sum1 += x
sumx = sum1

for l in range ( a+1, output-1 ) :
    sj = (ks * compare[l] )/(ks + kf) 
    x = compare[a] - (compare[l] - sj) #SkとSjは１対１の濃度で反応するから
    sum2 += x
sumx += sum2
#sumx = sum1+ sum2    
print(ks,sumx,rgj[a],yfj[a],t4)
yj = ks * sumx * rgj[a] * yfj[a] * t4   #d[Yj]/dt = ks[Sj][RGj][YFj]
print(kc,yj,repj[a],t5)
fluor_a = kc * yj * repj[a] * t5  #Y(a+1)がどのくらい光るかわかる d[Fluor]/dt = kc[Yj][Repj]
print(fluor_a,"a:",a)


all_fluor = [] #Yjに対応する蛍光する濃度を格納する
for other in range(0, a) :
    sj = (ks * compare[other] )/(ks + kf)   #SkがPair annihilationで消されている間にSignal restorationで反応する濃度をx(mol/l)とすると([Sk]-x)/kf =x/ksなのでx=ks[Sk]/(ks+kf)
    yj = ks* sj * rgj[other] * yfj[other] * t4  #d[Yj]/dt = ks[Sj][RGj][YFj]
    fluor = kc * yj * repj[other] * t5  #d[Fluor]/dt = kc[Yj][Repj]
    print(fluor,"other : ",other)
    all_fluor.append(fluor)
    
all_fluor.append(fluor_a)

for other in range(a+1, output) :
    sj = (ks * compare[other] )/(ks + kf)
    yj = ks* sj * rgj[other] * yfj[other] * t4
    fluor = kc * yj * repj[other] * t5
    print(fluor,"other2 : ",other)
    all_fluor.append(fluor)
print(all_fluor)    
```
